```python
x = np.arange(0,5,0.1)
y = x**3
y2 = x**2
plt.plot(x,y, color="blue", linewidth=5, label = "X^3")
plt.title("MatplotLib Demo")
plt.xlabel("X")
plt.ylabel("X^3")
plt.plot(x,y2, color = "black", linewidth=5, label = "X^2")
plt.legend()
# plt.axis([0,10,0,140]) # as per your use
plt.text(2,100,"Text")
plt.grid()
plt.show()    
```

```python
# Bar graph
import matplotlib.pyplot as plt

categories = ['Python', 'C++', 'Java', 'Go']
values = [80, 65, 90, 40]

plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title('Programming Language Popularity')
plt.xlabel('Languages')
plt.ylabel('Popularity (%)')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Histogram
import numpy as np

data = np.random.randn(1000)  # Normal distribution
plt.hist(data, bins=30, color='purple', edgecolor='white')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Pie chart
labels = ['Python', 'C++', 'Java', 'Go']
sizes = [40, 25, 20, 15]
colors = ['gold', 'lightblue', 'lightgreen', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Language Usage Share')
plt.axis('equal')  # Makes it a perfect circle
plt.show()
```

```python
# Bubble Plot
x = [10, 20, 30, 40]
y = [5, 15, 25, 35]
sizes = [100, 300, 500, 700]  # Bubble sizes

plt.scatter(x, y, s=sizes, alpha=0.5, c='teal', edgecolors='black')
# c='teal' -> tell the color, alpha = 0.5 -> the Transparency
plt.title('Bubble Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
